## Create folders for positive and negative data

```python
pos_pdb = []

for i, row in data.iterrows():
    if row['label'] == 1:
        pos_pdb.append(row['pdb'])
```

```python
neg_pdb = []

for i, row in data.iterrows():
    if row['label'] == 0:
        neg_pdb.append(row['pdb'])
```

```python
# importing the modules
import shutil
import os

# Providing the folder path
origin = '/home/palistha/disc/Lab/Certificates/data/20220817_0140880/'
pos_target = '/home/palistha/disc/Lab/Certificates/data/positive/'
neg_target = '/home/palistha/disc/Lab/Certificates/data/negative/'

# Fetching the list of all the files
files = os.listdir(origin)

# Fetching positive and negative pdbs to respective directories
for f in files:
    if f.split('.')[0] in pos_pdb: 
        shutil.move(origin + f, pos_target)
    elif f.split('.')[0] in neg_pdb: 
        shutil.move(origin + f, neg_target)
    else:
        pass
```

```python
count = 0
# Iterate directory
for path in os.listdir(pos_target):
    # check if current path is a file
    if os.path.isfile(os.path.join(pos_target, path)):
        count += 1
print('File count:', count)
```

```python
count = 0
# Iterate directory
for path in os.listdir(neg_target):
    # check if current path is a file
    if os.path.isfile(os.path.join(neg_target, path)):
        count += 1
print('File count:', count)
```
